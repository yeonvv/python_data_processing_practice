## 파이썬 기본 문법 복습


```python
x = 10  # 정수값 할당
x = "안녕하세요"  # 문자열값으로 재할당
print(x)
```

```python
# 다양한 타입의 변수 할당
a = 5  # 정수
b = 2.5  # 실수
c = True  # 불리언
d = "파이썬"  # 문자열
e = [1, 2, 3]  # 리스트
f = (4, 5, 6)  # 튜플
g = {"이름": "홍길동", "나이": 30}  # 딕셔너리
h = {1, 2, 2, 3}  # 집합 (중복된 2는 한 번만 저장됨, 순서 없음)

print(a, b, c, d)
print(e)
print(f[1])
print(g["이름"])
print(h)
```

```python
# 인덱싱과 슬라이싱
text = "Hello world"
print(text[0:5])
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])
```

```python
# 조건문
x = 10
if x > 0:
    print("x는 양수입니다.")
elif x == 0:
    print("x는 0입니다.")
else:
    print("x는 음수입니다.")
```

```python
# for 반복문
for i in range(5):
    print(i)

for fruit in ["사과", "배", "포도"]:
    print(fruit)
```

```python
# while 반복문
count = 5
while count > 0:
    print(count)
    count -= 1
print("발사!")
```

```python
# 함수의 정의와 호출
def add(a, b):
    result = a + b
    return result


def greet(name):
    print(f"안녕하세요, {name}님!")


sum_value = add(3, 5)
print(f"합계: {sum_value}")

greet("철수")
```
